# Discrete Event Simulation

## Simulating wait times in an emergency department 

- 3 triage.
- 1 physician.
- Each triage takes 3 minutes to complete.
- Each physician spends 5 minutes on every patient. 
- One patients arrive every 2 minutes.

```python
import pandas as pd
import simpy

class Emergency(object):
    import simpy
    def __init__(self, env, arrival, triage_cap, triage_duration, num_physicians, treatment_duration):
        
        self.env = env        
        self.triage_cap = triage_cap
        self.num_physicians = num_physicians
        self.triage_duration=triage_duration
        self.treatment_duration=treatment_duration
        self.arrival=arrival
    
            
    def patient(self):
        # Simulate patient arrival
        
        
        yield env.timeout(self.arrival)
        arrival_time.append(self.env.now)

        # Request one of the triage rooms
        with self.triage_cap.request() as req:
            yield req
            triage_start_time.append(env.now)

            
            
            yield self.env.timeout(self.triage_duration)
            
            triage_end_time.append(env.now)
            
        # Request for physician
        with self.num_physicians.request() as req:
            yield req
            

            treatment_start_time.append(env.now)
            yield self.env.timeout(self.treatment_duration)
            treatment_end_time.append(env.now)
```

```python
patient_number =[]
arrival_time =[]
triage_start_time, triage_end_time=[],[]
treatment_start_time, treatment_end_time=[],[]

'''************************************************'''

env = simpy.Environment()
TRIAGE_CAP = simpy.Resource(env, capacity=3)
NUM_PHYSICIANS = simpy.Resource(env, capacity=1)
TRIAGE_DURATION=3
TREATMENT_DURATION=5
'''************************************************'''


for i in range(0,20):
    patient_number.append(i+1)
    emerg = Emergency(env, arrival=i*2, triage_cap=TRIAGE_CAP, triage_duration=TRIAGE_DURATION,
                      num_physicians=NUM_PHYSICIANS, treatment_duration=TREATMENT_DURATION)
    env.process(emerg.patient())

env.run()
```

```python
simulation = pd.DataFrame(
{'patient number':patient_number,
 'arrival_time':arrival_time,
 'triage_start_time':triage_start_time,
 'triage_end_time':triage_end_time,
 'treatment_start_time':treatment_start_time,
 'treatment_end_time':treatment_end_time
})

simulation['wait_time']=simulation['treatment_start_time']-simulation['triage_start_time']

simulation
```
